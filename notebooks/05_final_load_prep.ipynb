# 05: Final Data Preparation for Tableau

This notebook validates the final cleaned dataset and confirms it is ready for the Tableau Public dashboard build.

## Setup

```python
import pandas as pd
```

## Load Final Dataset

```python
df = pd.read_csv('../data/processed/cleaned_diabetic_data.csv')
print(f'Final Dataset: {df.shape[0]:,} rows x {df.shape[1]} columns')
```

## Quality Assurance Checks

```python
print('=== Null Values ===')
print(f'Total nulls: {df.isnull().sum().sum()}')

print('\n=== Target Variable Distribution ===')
print(f"Readmitted Under 30 Days: {df['Readmitted Under 30 Days'].sum():,} ({df['Readmitted Under 30 Days'].mean()*100:.2f}%)")
print(f"Readmitted After 30 Days: {df['Readmitted After 30 Days'].sum():,} ({df['Readmitted After 30 Days'].mean()*100:.2f}%)")
print(f"Not Readmitted: {df['Not Readmitted'].sum():,} ({df['Not Readmitted'].mean()*100:.2f}%)")

print('\n=== Column Data Types ===')
print(df.dtypes.value_counts())

print('\n=== KPI Summary ===')
print(f"30-Day Readmission Rate: {df['Readmitted Under 30 Days'].mean()*100:.2f}%")
print(f"Average Length of Stay: {df['Days in Hospital'].mean():.2f} days")
```

## Column Inventory for Tableau

```python
for i, col in enumerate(df.columns, 1):
    print(f'{i:2d}. {col:40s} | {str(df[col].dtype):8s} | {df[col].nunique():6,} unique values')
```
